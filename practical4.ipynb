```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, confusion_matrix, precision_score, recall_score, accuracy_score
```

```python
df = pd.read_csv("C:\\Users\\jit16\\Downloads\\DSBDA from bhushan\\Boston.csv")
```

```python
df
```

```python
df.columns
```

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS',
       'RAD', 'TAX', 'PTRATIO', 'BLACK', 'LSTAT']]
```

```python
y = df['MEDV']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.score(x, y)
```
