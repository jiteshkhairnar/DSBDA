```scala
object LogMapReduce {
  def main(args: Array[String]): Unit = {

    val logs = scala.io.Source.fromFile("/home/student/log.txt").getLines().toList

    val mapped = logs.flatMap { line =>
      val parts = line.split(" ")
      if (parts.nonEmpty) Some((parts(0), 1)) else None
    }

    val reduced = mapped
      .groupBy(_._1)
      .map { case (key, values) =>
        (key, values.map(_._2).sum)
      }

    reduced.foreach(println)
  }
}
```

```scala
LogMapReduce.main(Array())
```
