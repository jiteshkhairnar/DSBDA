```python
sentence1 = "i will walk 500 miles and i would walk 500 more. just to be the man who walks a thousand miles to fall down at your door!"
sentence2 = "I played the play playfully as the players were playing in the play with playfullness"
```

```python
from nltk import word_tokenize,sent_tokenize
print('Tokenize words : ',word_tokenize(sentence1))
print('Tokenize sentence : ',sent_tokenize(sentence1))
```

```python
from nltk import pos_tag
token=word_tokenize(sentence1)+word_tokenize(sentence2)
tag=pos_tag(token)
print("Tagged POS : ",tag)
```

```python
from nltk.corpus import stopwords
stop_word=stopwords.words('english')
token=word_tokenize(sentence1)
clean_token=[]
for word in token:
    if word not in stop_word:
        clean_token.append(word)
print('Unclean version : ',token)
print('\ncleand version : ',clean_token)
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
token=word_tokenize(sentence2)
stemmed=[stemmer.stem(word) for word in token]
print(' '.join(stemmed))
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
token=word_tokenize(sentence2)
lem_output=[lem.lemmatize(word) for word in token]
print(' '.join(lem_output))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = [ "i will walk 500 miles and i would walk 500 more. just to be the man who walks a thousand miles to fall down at your door!"]
vectorizer = TfidfVectorizer()
print(vectorizer)
tfidf_matrix = vectorizer.fit_transform(documents)
print("Words:", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix)
```
